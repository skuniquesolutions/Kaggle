```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

step 1 - Data Preparation:

```python
df1 = pd.read_csv('train.csv')
```

```python
df1.head()
```

step 2 - Identify Missing Values and Data Visulatization

a. To identify missing values for all columns in the DataFrame:

```python
missing_value = df1.isnull().sum()
print(missing_value)
```

```python
missing_value.describe()
```

b. To identify the percentage of missing values for each column:

```python
missing_percentage = (df1.isnull().sum()/len(df1))*100
print(missing_percentage)
```

c. To check for missing values in a specific column (e.g., 'column_name'):

```python
#c. To check for missing values in a specific column (e.g., 'column_name'):

for i in df1.columns:
    if df1[i].isnull().sum() > 1:
        print(f"{i}:-",df1[i].isnull().sum())
```

```python


import seaborn as sns
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df1.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Data Heatmap')
plt.show()
```

```python
#To get a summary of missing values statistics:

missing_summary = df1.describe(include='all')
print(missing_summary)

#You can identify missing values by looking at the 'count' row.
```

```python
#Bar chart
plt.figure(figsize=(12, 6))
missing_values = df1.isnull().sum()
missing_values.plot(kind='bar', alpha=0.7)
plt.show()
```

step 3- Summary Statistics

```python
summary = df1.describe()
print(summary)
```

```python
# Generate summary statistics for all columns (including non-numeric) and customize percentiles
custom_summary = df1.describe(include='all', percentiles=[0.25, 0.5, 0.75])

# Print the custom summary statistics
print(custom_summary)
```

```python
#2) Correlation Analysis

correlation_martix = df1.corr()
print(correlation_martix)
```

```python
# Set the figure size
plt.figure(figsize=(10, 8))

sns.heatmap(correlation_martix, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

Interpret the results:

Positive values indicate a positive correlation (as one variable increases, the other tends to increase).
Negative values indicate a negative correlation (as one variable increases, the other tends to decrease).
Values close to 0 suggest a weak or no linear relationship.
