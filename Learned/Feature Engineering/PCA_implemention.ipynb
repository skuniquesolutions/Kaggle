```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn import datasets
```

```python
data = datasets.load_breast_cancer()
data.keys()
```

```python
print(data.DESCR)
```

```python
#convert into dataframe

df = pd.DataFrame(data.data, columns=data.feature_names)

df.head()
```

```python
len(df.columns)
```

```python
#Standardization
from sklearn.preprocessing import StandardScaler

scale = StandardScaler()

scale.fit(df)

df.head()
```

```python
scaled_data = scale.transform(df) 

scaled_data, type(scaled_data)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2) #extract the top 2 feature out of all features

data_pca = pca.fit_transform(scaled_data)

data_pca
```

```python
pca.explained_variance_ #maxmium variance pc1 and pc2
```

```python
data_pca[:,1]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.scatter(data_pca[:,0], data_pca[:,1], c = data.target)

plt.xlabel('PCA_1')
plt.ylabel('PCA_2')
plt.show()
```
